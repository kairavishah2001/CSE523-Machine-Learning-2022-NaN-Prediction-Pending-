```python
import plotly.graph_objects as go
import numpy as np

x_lgbm = np.array([3, 5, 10, 13])
y_lgbm = np.array([0.58, 0.66, 0.69, 0.72])

x_xgb = np.array([3, 5, 10, 13])
y_xgb = np.array([0.52, 0.67, 0.67, 0.70])

x_cb3 = np.array([3, 5, 10, 13])
y_cb3 = np.array([0.52, 0.53, 0.52, 0.54])

x_cb5 = np.array([3, 5, 10, 13])
y_cb5 = np.array([0.54, 0.54, 0.57, 0.56])

x_cb7 = np.array([3, 5, 10, 13])
y_cb7 = np.array([0.54, 0.55, 0.52, 0.51])

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_lgbm, y=y_lgbm, name="LGBM",
                    line_shape='linear'))
fig.add_trace(go.Scatter(x=x_xgb, y=y_xgb, name="XGB",
                    line_shape='linear'))
fig.add_trace(go.Scatter(x=x_cb3, y=y_cb3, name="CB3",
                    line_shape='linear'))
fig.add_trace(go.Scatter(x=x_cb5, y=y_cb5, name="CB5",
                    line_shape='linear'))
fig.add_trace(go.Scatter(x=x_cb7, y=y_cb7, name="CB7",
                    line_shape='linear'))
fig.update_traces(hoverinfo='text+name', mode='lines+markers')
fig.update_layout(legend=dict(y=0.5, traceorder='reversed', font_size=16))

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x = np.array([3, 5, 10, 13])
y = np.array([0.58, 0.66, 0.69, 0.72])

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x, 
        y=y, 
        name="linear",
        line_shape='linear')
)

fig.update_traces(hoverinfo='text+name', mode='lines+markers')
fig.update_layout(legend=dict(y=0.5, traceorder='reversed', font_size=16), title="Light Gradient Boosting Model",font=dict(size= 15))

fig.show()
```
